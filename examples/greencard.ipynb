```python
import torch

import sys
sys.path.append('..')

from rsa import RSA
from plot_utils import plot_dist
```

```python
gc = RSA("../sample_configs/greencard_3_worlds.json")
```

```python
utterance = "not green card"
qud = "need visa?"
context = torch.tensor([True, True, True])
world = torch.tensor([True, False, False])
```

```python
plot_dist(
    dist=gc.literal_listener(utterance, context),
    title=f"Lit Listener dist over worlds with utterance '{utterance} and qud '{qud}'",
    names=gc.get_world_labels()
)
```

```python
plot_dist(
    dist=gc.speaker(world, qud, context),
    title=f"Speaker dist over utterances with qud '{qud}'",
    names=gc.get_utterance_labels()
)
```

```python
plot_dist(
    gc.pragmatic_listener(utterance, qud),
    title=f"Pragmatic listener dist over worlds with utterance '{utterance} and qud '{qud}'",
    names=gc.get_world_labels()
)
```

```python
plot_dist(
    gc.pragmatic_listener(utterance, qud, output_type="context"),
    title=f"Pragmatic listener dist over contexts with utterance '{utterance} and qud '{qud}'",
    names=gc.get_context_labels()
)
```
