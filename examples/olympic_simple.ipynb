```python
import torch

import sys
sys.path.append('..')

from rsa import RSA
from plot_utils import plot_dist
```

```python
fgs = rsa.RSA("../sample_configs/olympic_sprinter_simple.json")
```

```python
utterance = "Olympic sprinter"
qud = "profession"
context = torch.tensor([True, True, True, True])
world = torch.tensor([False, False, True, False])
```

```python
plot_dist(
    dist=fgs.literal_listener(utterance, context),
    title=f"Lit Listener dist over worlds with utterance '{utterance} and qud '{qud}'",
    names=fgs.get_world_labels()
    )
```

```python
plot_dist(
    dist=fgs.speaker(world, qud, context),
    title=f"Speaker dist over utterances with qud '{qud}'",
    names=fgs.get_utterance_labels()
    )
```

```python
plot_dist(
    fgs.pragmatic_listener(utterance, qud),
    title=f"Pragmatic listener dist over worlds with utterance '{utterance} and qud '{qud}'",
    names=fgs.get_world_labels()
    )
```

```python
plot_dist(
    fgs.pragmatic_listener(utterance, qud, output_type="context"),
    title=f"Pragmatic listener dist over contexts with utterance '{utterance} and qud '{qud}'",
    names=fgs.get_context_labels()
    )
```
